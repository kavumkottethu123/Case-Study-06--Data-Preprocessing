## Filling the missing values

```python
titanic_data.columns
```

```python
titanic_data.isnull().sum()
```

### Insights:
* ‘Survived’ is the target variable, which we will predict once our preprocessing of our data is done. So, we retain that column.
* Only the columns such as ‘Age’, ‘Cabin’ and ‘Embarked’ has missing values.
* ‘PassengerId’, ‘Name’ and ‘Ticket’ doesn’t add much value in predicting the target variable.
* ‘ParCh’(Parent/Children) and ‘SibSp’(Siblings/Spouse) details are related to family.
* ‘Sex’, 'Embarked' are the categorical data that needs to be encoded to numerical values.

```python
#Ignore the “Cabin” column since 70% of that column is missing. 
titanic_data.drop(['Cabin'],1,inplace=True) #Dropping cabin column
```

```python
#Only two rows are missing the values for Embarked column.
# Returns count for each category
titanic_data['Embarked'].value_counts()
```

```python
# Fills null values with 'S'-most common occurence
common = 'S'
titanic_data['Embarked']=titanic_data['Embarked'].fillna('S')
# Checking the no of null values now
titanic_data['Embarked'].isnull().sum()
```

```python
# Filling the missing values in the age column
#we are going to use the median of that column.
titanic_data['Age'] = titanic_data['Age'].fillna((titanic_data['Age'].median()))
```

```python
titanic_data.isnull().sum()
```

Now all the missing values are filled.

```python
#Encoding categorical features
titanic_data['Embarked'] = titanic_data['Embarked'].map({'C':0, 'Q':1, 'S':2})
titanic_data['Sex'] = titanic_data['Sex'].map({'male':0, 'female':1})
```

```python
#Dropping of columns
titanic_data.drop(['Name','Ticket'], 1, inplace=True)

#‘ParCh’ and ‘SibSp’ are the details related to family size, so let’s derive a new column named ‘Size of the Family’.
titanic_data['FamilySize'] = titanic_data['SibSp'] + titanic_data['Parch']+1
```

#### Model Development

```python
titanic_data.columns
```

```python
#Splitting dataset into dependent and independent variables
y=titanic_data['Survived']
x=titanic_data.drop('Survived',axis=1)
```

```python
#Splitting into trainset and test set
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,random_state=42,test_size=0.2)
```

```python
#Linear Regression model
from sklearn import linear_model
lr=linear_model.LinearRegression()
model=lr.fit(x_train,y_train)
predictions=model.predict(x_test)
```

#### Model Evaluation

```python
from sklearn.metrics import mean_squared_error
print('MSE IS:',mean_squared_error(y_test,predictions))
```

```python
from sklearn.metrics import r2_score
print('R squared value is:',r2_score(y_test,predictions))
```

```python
x=titanic_data.drop('Survived',axis=1)
```

```python
from sklearn import preprocessing
```

```python
min_max=preprocessing.MinMaxScaler(feature_range=(0,1))
```

```python
x=min_max.fit_transform(x)
```

```python
x=pd.DataFrame(x)
```

```python
x.describe()
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,random_state=42,test_size=0.2)
lr=linear_model.LinearRegression()
model=lr.fit(x_train,y_train)
predictions=model.predict(x_test)
print('MSE IS:',mean_squared_error(y_test,predictions))
print('R squared value is:',r2_score(y_test,predictions))
```
